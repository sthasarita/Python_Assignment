```python
# Employee Record Reader
id = input("Enter employee id:")
try:
    id = int(id)
    with open("employees.txt", "r") as f:
        for line in f:
            rec = line.strip().split(",")
            found = False
            
            if id == int(rec[0]):
                
                found = True
                print("Employee Found")
                print(f"ID: {rec[0]}")
                print(f"Name: {rec[1]}")
                print(f"Salary: {rec[2]}")
        if not found:
                print("Employee Not Found")
except FileNotFoundError:
    print("Error: Employee file not found")
except ValueError:
    print("Please enter a valid numeric employee ID.")
except IndexError:
    print("Corrupted employee record found.")
finally:
    print("Search completed.")
    
```

```python
## Student Division Calculator
print("======== Student Division Calculator=======")
print("1. Calculate Percentage")
print("2. Exit")
while True:
    try:
        choice = int(input("Enter your choice:"))
        
        if choice == 1:
            name = input("Enter Student Name:")
            t_marks = int(input("Enter total marks:"))
            o_marks = int(input("Enter obtained marks:"))
            if t_marks < o_marks:
                print("Total marks can't be less than obtained marks.")
            else:
                percentage = (o_marks / t_marks) * 100
            print(f"Student Name : {name}")
            print(f"Percentage : {percentage:.2f}%")
        elif choice == 2:
            print("Exiting...")
            break
        else:
            print("Invalid input.")

    except ValueError:
        print("Invalid input. Please enter numbers only.")
    except ZeroDivisionError:
        print("Error occurred! Can't divide by zero. Total Marks cannot be zero")
    except KeyboardInterrupt:
        print("\n Program interrupted by user.")
        break
    else:
        print("Calculation Successful. As no exception occured in the try block so else block runs.")
    finally:
        print("Thank you for using the calculator.")
        
        
```
