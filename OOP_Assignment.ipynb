```python
# Book Management System
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
    def display_details(self):
        print(f"Book Title : {self.title}")
        print(f"Author : {self.author}")
        print(f"Price : {self.price}")
    def discount(self):
        price_after_dis = self.price - 0.1 * self.price
        print(f"Original Price : Rs.{self.price}")
        print(f"Price after 10% discount : Rs.{price_after_dis:.2f}")
    def __str__(self):
        return f"Book(Title= {self.title}, Author={self.author}, Price={self.price})"
    def is_expensive(self):
        if self.price > 1000:
            print("This is an expensive book.")
        else:
            print("This is a budget-friendly book.")
        

book1 = Book("Python Basics", "Sarita Shrestha", 760)
book2 = Book("Muna Madan", "Laxmi Prasad Devkota", 400)    
book1.display_details()
print()
book1.discount()
print()
print(book1)
print()
book2.display_details()
print()
book2.discount()
print()
print(book2)
```

```python
book1.is_expensive()
```

```python
book2.is_expensive()
```

```python
book3 = Book("python", "Ram", 1200)
book3.is_expensive()
```

```python
#Inheritance (Vehicle Management System)
class Vehicle():
    def __init__(self, brand , model):
        self.brand = brand
        self.model = model
    def display_vehicle(self):
        print(f"Brand : {self.brand}")
        print(f"Model : {self.model}")
        
class Car(Vehicle):
    def __init__(self, brand, model, price):
        
        super().__init__(brand,model)
        self.price = price
    def display_price(self):
        print(f"price : {self.price}")
    def __str__(self):
        return f"Car(Brand = {self.brand} , Model = {self.model}, Price = {self.price})"

car1 = Car("Toyota", "Corolla", 4500000)

car2 = Car("Hyundai", "Creta", 5200000)

        
print("Printing the Car 1 details:")
print(car1)
car1.display_vehicle()
car1.display_price()
print()



print("Printing the Car 2 details:")
print(car2)
car2.display_vehicle()
car2.display_price()
        
```

```python
# Abstraction (Employee Salary System)
from abc import ABC, abstractmethod

class Employee(ABC):
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    @abstractmethod
    def work():
        pass

    def display(self):
        print(f"Employee Name:{self.name}")
        print(f"Salary : {self.salary}")

class SoftwareEngineer(Employee):
    def work(self):
        print(f"{self.name} is developing software.")
    def __str__(self):
        return f"SoftwareEngineer (Name = {self.name} , Salary = {self.salary})"

class Teacher(Employee):
    def work(self):
        print(f"{self.name} is teaching students.")
    def __str__(self):
        return f"Teacher (Name = {self.name} , Salary = {self.salary})"

print("Employee 1 details:")

emp1 = SoftwareEngineer("Ram", 50000)
print(emp1)
print(emp1.display())
print(emp1.work())

print()

print("Employee 2 details:")
emp2 = Teacher("Sita", 35000)
print(emp2)
print(emp2.display())
print(emp2.work())

        
```
