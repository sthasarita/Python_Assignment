```python
# Book Management System
class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price
    def display_details(self):
        print(f"Book Title : {self.title}")
        print(f"Author : {self.author}")
        print(f"Price : {self.price}")
    def discount(self):
        price_after_dis = self.price - 0.1 * self.price
        print(f"Original Price : Rs.{self.price}")
        print(f"Price after 10% discount : Rs.{price_after_dis:.2f}")
    def __str__(self):
        return f"Book(Title= {self.title}, Author={self.author}, Price={self.price})"

book1 = Book("Python Basics", "Sarita Shrestha", 760)
book2 = Book("Muna Madan", "Laxmi Prasad Devkota", 400)    
book1.display_details()
print()
book1.discount()
print()
print(book1)
print()
book2.display_details()
print()
book2.discount()
print()
print(book2)
```
