```python
#Student Record File Management
while True:
    print("\n========== Student Record System ==========")
    print("1. Add Student")
    print("2. Display All Students")
    print("3. Search Student")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    # ------------------ ADD STUDENT ------------------
    if choice == 1:
        name = input("Enter Name: ")
        age = int(input("Enter Age: "))
        marks = int(input("Enter Marks: "))

        with open("students.txt", "a") as f:
            f.write(f"{name},{age},{marks}\n")

        print("Student added successfully!")

    # ---------------- DISPLAY ALL --------------------
    elif choice == 2:
        try:
            with open("students.txt", "r") as f:
                print("\n----- Student Records -----")

                for line in f:
                    line = line.strip()
                    record = line.split(",")

                    print(f"Name  : {record[0]}")
                    print(f"Age   : {int(record[1])}")
                    print(f"Marks : {int(record[2])}")
                    print()

        except FileNotFoundError:
            print("No student records found.")

    # ---------------- SEARCH STUDENT -----------------
    elif choice == 3:
        search_name = input("Enter student name: ")

        found = False

        try:
            with open("students.txt", "r") as f:

                for line in f:
                    record = line.strip().split(",")

                    if record[0].lower() == search_name.lower():
                        found = True

                        print("\nStudent Found")
                        print(f"Name  : {record[0]}")
                        print(f"Age   : {int(record[1])}")
                        print(f"Marks : {int(record[2])}")
                        break

            if not found:
                print("Student not found.")

        except FileNotFoundError:
            print("No student records found.")

    # ---------------- EXIT ------------------
    elif choice == 4:
        print("Exiting...")
        break

    else:
        print("Invalid choice!")
```

```python
#Employee Salary Report
highest_salary = 0
highest_employee = ""
highest_department = ""

total_salary = 0
employee_count = 0

department_count = {}

with open("employees.txt", "r") as f:

    print("------ Employee Details ------")

    for line in f:

        record = line.strip().split(",")

        emp_id = record[0]
        name = record[1]
        salary = int(record[2])
        department = record[3]

        # A. Display all employees
        print(f"Employee ID : {emp_id}")
        print(f"Name        : {name}")
        print(f"Salary      : {salary}")
        print(f"Department  : {department}")
        print()

        # B. Highest salary
        if salary > highest_salary:
            highest_salary = salary
            highest_employee = name
            highest_department = department

        # C. Average salary
        total_salary += salary
        employee_count += 1

        # D. Count employees in each department
        if department in department_count:
            department_count[department] += 1
        else:
            department_count[department] = 1

# B
print("------ Highest Salary ------")
print(f"Name       : {highest_employee}")
print(f"Salary     : {highest_salary}")
print(f"Department : {highest_department}")

# C
average = total_salary / employee_count

print("\nAverage Salary =", average)

# D
print("\n------ Department Count ------")

for department, count in department_count.items():
    print(f"{department} -> {count}")

# E
with open("high_salary.txt", "w") as f:

    with open("employees.txt", "r") as file:

        for line in file:

            record = line.strip().split(",")

            name = record[1]
            salary = int(record[2])

            if salary > 60000:
                f.write(f"{name},{salary}\n")
```
