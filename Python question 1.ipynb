**Q1: Variable Swapping**
Swap the values of two variables `a` and `b` without using a third temporary variable.

```python
def swap_values(a, b):
    # Write your code here
    a,b = b,a
    return a,b
    pass

# Validation
x, y = 5, 10
x, y = swap_values(x, y)
assert (x, y) == (10, 5)
print("Q1 Success!")
```

**Q2: Dynamic Type Checking**
Write a function that returns `True` if a variable is either an `int` or a `float`, and `False` otherwise.

```python
def is_numeric(val):
    # Write your code here
    if isinstance(val, (int,float)):
        return True
    else:
        return False

# Validation
assert is_numeric(5) == True
assert is_numeric(5.5) == True
assert is_numeric("5") == False
print("Q2 Success!")
```

**Q3: String Slicing Logic**
Return a string consisting of the first 2 and last 2 characters of a given string. If the string length is less than 2, return an empty string.

```python
def string_ends(s):
    # Write your code here
    if len(s) >= 2:
        return s[:2] + s[-2:]
    else:
        return ""

# Validation
assert string_ends("Python") == "Pyon"
assert string_ends("A") == ""
print("Q3 Success!")
```

**Q6: The FizzBuzz Logic**
Return "Fizz" if `n` is divisible by 3, "Buzz" if by 5, "FizzBuzz" if by both, and the number itself as a string otherwise.

```python
def fizzbuzz(n):
    # Write your code here
    if n % 3 == 0 and n % 5 == 0:
        return "FizzBuzz"
    elif n % 5 == 0:
        return "Buzz"
    elif n % 3 == 0 :
        return "Fizz"
    else:
        return str(n)

# Validation
assert fizzbuzz(15) == "FizzBuzz"
assert fizzbuzz(3) == "Fizz"
assert fizzbuzz(7) == "7"
print("Q6 Success!")
```

**Q7: Prime Number Check**
Write a function to determine if a number `n` is prime.

```python
def is_prime(n):
    # Write your code here
    if n<=1:
        return False
    prime = True
    for i in range(2, n):
        if n % i == 0:
            prime = False
            break
    if prime:
        return True
    else:
        return False
    
        

# Validation
assert is_prime(11) == True
assert is_prime(4) == False
print("Q7 Success!")
```

**Q8: Range Summation**
Sum all even numbers between 1 and `n` (inclusive) using a `for` loop.

```python
def sum_evens(n):
    # Write your code here
    total = 0
    for i in range(1, n+1):
        if i % 2 == 0:
            total += i
    return total

# Validation
assert sum_evens(10) == 30
print("Q8 Success!")
```

**Q9: Palindrome Logic**
Check if a string is a palindrome, ignoring case.

```python
def is_palindrome(s):
    # Write your code here
    l = list(s)
    l.reverse()
    r_s = ""
    for char in l:
        r_s += char
    if s == r_s:
        return True
    else:
        return False

    # alternative way
    # clean_s = s.lower()
    # return clean_s == clean_s[::-1]

# Validation
assert is_palindrome("RacecaR") == True
assert is_palindrome("hello") == False
print("Q9 Success!")
```

**Q11: Factorial via Loop**
Calculate `n!` (factorial) using a loop (not recursion).

```python
def factorial_loop(n):
    # Write your code here
    fact = 1
    for i in range(2, n+1):
        fact = fact * i 
    return fact

# Validation
assert factorial_loop(5) == 120
print("Q11 Success!")
```

**Q14: Lambda Functions**
Use a lambda function to sort a list of tuples `[(1, 10), (2, 5), (3, 8)]` by the second element.

```python
def sort_tuples(data):
    # Write your code here (return the sorted list)
    data = sorted(data , key = lambda x : x[1])
    return data
    

# Validation
assert sort_tuples([(1, 10), (2, 5), (3, 8)]) == [(2, 5), (3, 8), (1, 10)]
print("Q14 Success!")
```

**Q15: Recursive Fibonacci**
Write a recursive function to find the n-th Fibonacci number.

```python
def fib_recursive(n):
    # Write your code here
    if n == 1:
        return 1
    elif n== 2:
        return 1
    else: 
        return fib_recursive(n-1) + fib_recursive(n-2)

# Validation
assert fib_recursive(6) == 8
print("Q15 Success!")
```

**Q18: List Comprehension**
Given a list of numbers, return a new list containing the squares of only the odd numbers.

```python
def squares_of_odds(nums):
    # Write your code here
    sq_odd_nums = [num ** 2 for num in nums if num % 2 != 0]
    return sq_odd_nums

# Validation
assert squares_of_odds([1, 2, 3, 4, 5]) == [1, 9, 25]
print("Q18 Success!")
```

**Q21: Tuple Unpacking**
Given a list of tuples `[('Alice', 25), ('Bob', 30)]`, return a list of just the names.

```python
def extract_names(data):
    # Write your code here
    name_list = []
    for d in data:
        name_list.append(d[0])
    return name_list

# Validation
assert extract_names([('Alice', 25), ('Bob', 30)]) == ['Alice', 'Bob']
print("Q21 Success!")
```

**Q22: Frequency Map**
Return a dictionary showing the frequency of each character in a string.

```python
def char_frequency(s):
    # Write your code here
    dict_count= {}
    for char in s:
        if char in dict_count:
            dict_count[char] += 1
        else:
            dict_count[char] = 1
    
    return dict_count

# Validation
assert char_frequency("apple")['p'] == 2

print("Q22 Success!")
```

**Q23: List Rotation**
Rotate a list to the right by `k` steps.

```python
def rotate_list(lst, k):
    # Write your code here
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

# Validation
assert rotate_list([1, 2, 3, 4], 1) == [4, 1, 2, 3]
print("Q23 Success!")
```

**Q24: Safe Division**
Write a function that divides `a` by `b`. Use `try-except` to handle `ZeroDivisionError` and return `None` if it occurs.

```python
def safe_divide(a, b):
    # Write your code here
    try:
        return a/b
    except ZeroDivisionError:
        return None

# Validation
assert safe_divide(10, 2) == 5
assert safe_divide(10, 0) == None
print("Q24 Success!")
```

**Q25: File Content Read**
Write a function to create a file named `test.txt`, write "Hello Python" to it, and then read it back.

```python
def file_io_test():
    # Write your code here
    with open("test.txt", "w+") as f:
        f.write("Hello Python")
        f.seek(0)
        return f.read()


# Validation
assert file_io_test() == "Hello Python"
print("Q25 Success!")
```

**Q26: Custom Exceptions**
Create a custom exception `NegativeNumberError` and raise it if a function `check_positive` receives a negative number.

```python
# Define exception here
# The raise keyword creates an exception and immediately stops the current flow of execution.
class NegativeNumberError():
    pass
def check_positive(n):
    if n > 0:
        print(f"The number {n} is positive.")
    elif n < 0:
        raise NegativeNumberError("Negative number is not allowed.")
    else:
        print(f"Number {n} is zero.")


# Validation
try:
    check_positive(-5)
except Exception as e:
    print(f"Caught: {type(e).__name__}")
```

Given a multiline string representing CSV data (Name,Age), parse it into a list of dictionaries.

```python
import csv
csv_data = "Name,Age\nAlice,25\nBob,30"

def parse_csv(data):
    # Write your code here
    # DictReader automatically parses headers and rows in optimized C code
    reader = csv.DictReader(data.splitlines())
    return list(reader)
    
 
# Validation
result = parse_csv(csv_data)
print(result)
assert result[0]['Name'] == 'Alice'
print("Q27 Success!")
```
