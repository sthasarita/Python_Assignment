```python
# ============================================================
# Employee Data Processing Using Generators (Data Science Style)
# ============================================================

employees = [
    {"id": 101, "name": "Ram", "salary": 45000},
    {"id": 102, "name": "Hari", "salary": 85000},
    {"id": 103, "name": "Sita", "salary": 72000},
    {"id": 104, "name": "Gita", "salary": 65000},
    {"id": 105, "name": "Shyam", "salary": 95000},
]

# ------------------------------------------------------------
# Step 1: Generator that yields one employee at a time
# ------------------------------------------------------------

def employee_generator(data):
    for employee in data:
        yield employee


# ------------------------------------------------------------
# Step 2: Generator that filters employees with salary > 70000
# ------------------------------------------------------------

def high_salary(data):
    for employee in data:
        if employee["salary"] > 70000:
            yield employee


# ------------------------------------------------------------
# Step 3: Generator that converts names to uppercase
# ------------------------------------------------------------

def uppercase_name(data):
    for employee in data:
        # Make a copy so the original data is not modified
        new_employee = employee.copy()
        new_employee["name"] = new_employee["name"].upper()
        yield new_employee


# ============================================================
# Demonstrating next()
# ============================================================

print("Using next() on employee_generator:\n")

gen = employee_generator(employees)

print(next(gen))
print(next(gen))

print("\nRemaining employees using the same generator:\n")

for employee in gen:
    print(employee)


# ============================================================
# Generator Pipeline (Data Science Style)
# ============================================================

print("\nFinal Processed Employees:\n")

gen = employee_generator(employees)
gen = high_salary(gen)
gen = uppercase_name(gen)

for employee in gen:
    print(employee)


# ============================================================
# Demonstrating Generator Exhaustion
# ============================================================

print("\nGenerator Exhaustion Example:\n")

gen = employee_generator(employees)

for employee in gen:
    print(employee)

print("\nTrying next() after generator is exhausted:")

try:
    print(next(gen))
except StopIteration:
    print("StopIteration: Generator is exhausted.")
```
