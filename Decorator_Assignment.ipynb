```python
# closure
# Write a function make_power(exponent) that returns another function. The returned function should raise any given number to the stored exponent.

def make_power(exponent):
    def calculate_power(number):
        return number**exponent
    return calculate_power

square = make_power(2)
print(square(3))
cube = make_power(3)
print(cube(5))
```

```python
# First-Class Functions
# Question

# Write a function operate(func, a, b) that accepts another function and two numbers.

# Create two functions:

# add(a, b)
# subtract(a, b)

# Then call

# operate(add, 10, 5)
# operate(subtract, 10, 5)

def operate(func, a, b):
    return func(a,b)

def add(a,b):
    return a+b
def subtract(a,b):
    return a-b

print(operate(add, 5, 11))
print(operate(subtract, 5, 11))
```

```python
# Simple Decorator
# Question

# Create a decorator named start_end that prints

# Program Started

# before a function executes and

# Program Finished

# after it executes.

# Decorate a function

# greet(name)


def start_end(func):
    def wrapper(name):
        print("Program Started")
        result = func(name)
        print("Program finished")
        return result
    return wrapper
@start_end
def greet(name):
    print(f"Hello! {name}")
greet("saru")
        
```

```python
# Without using @, decorate this function manually.

# def greet():
#     print("Hello")

# using

# simple_decorator

def simple_decorator(func):
    def wrapper():
        print("Program Started")
        result = func()
        print("Program finished")
        return result
    return wrapper
def greet():
    print("Hello")
greet = simple_decorator(greet)
greet()
```

```python
# Create a decorator show_arguments that prints all positional and keyword arguments before calling the original function.
def show_arguments(func):
    def wrapper(*args, **kwargs):
        
        print("Program Started")
        print(f"Positional arguments: {args}")
        print(f"Keyword arguments: {kwargs}")
        result = func(*args, **kwargs)
        print("Program finished")
        print(f"Positional arguments: {args}")
        print(f"Keyword arguments: {kwargs}")
        return result
    return wrapper
    
@show_arguments
def student(name, age, city= "Kathmandu"):
    print(f"Your name is {name}")
    print(f"Your age is {age}")
    print(f"Your city is {city}")

student("saru", 12, city= "Pokhara")
```

```python
# Decorator That Takes Arguments
# Question

# Create a decorator

# @repeat(3)

# that executes the decorated function exactly 3 times.

def repeat(times):
    def s_decorator(func):
        def wrapper():
            for i in range(1, times + 1):
                func()
        return wrapper
    return s_decorator
@repeat(3)
def hello():
    print("hello!")
hello()


        
```

```python
# Create two decorators uppercase add_exclamation Decorate message() so the output becomes WELCOME!!! using both decorators together.
def uppercase(func):
    def wrapper():
        result = func()
        return result.upper()
    return wrapper

def add_exclamation(func):
    def wrapper():
        result = func()
        return result+"!!!"
    return wrapper


@add_exclamation
@uppercase
def message():
    return "welcome"

print(message())

    
```

```python
# Create a class decorator

# TrackCalls

# that counts how many times a function has been called.

class TrackCalls():
    def __init__(self, func):
        self.func = func 
        self.count = 0

    def __call__(self):
        self.count += 1
        print(f"Call {self.count} of {self.func.__name__}")
        print(self.func())

@TrackCalls
def hello():
    return "hello"

hello()
hello()
hello()
print(f"Total counts = {hello.count}")
```

```python
# example of static method
class Calculator():
    @staticmethod
    def multiply(a,b):
        return a * b
print(Calculator.multiply(3,4))
obj_cal = Calculator()
print(obj_cal.multiply(3,4))
```

```python
# example of classmethod
class Student():
    school = "Techaxis"

    @classmethod
    def change_school(cls,new_name):
        cls.school = new_name

s1 = Student()
print(s1.school)
s1.change_school("Samata")
print(s1.school)
```

```python
# example of property method
class Circle:
    def __init__(self, radius):
        self.radius = radius
    @property
    def area(self):
        return 3.14 * self.radius * self.radius

c = Circle(5)
print(c.area)
```
